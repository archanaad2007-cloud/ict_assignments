```python
for i in range(1,101):
  if i%3==0 and i%5==0:
    print("FizzBuzz")
  elif i%3==0:
    print("Fizz")
  elif i%5==0:
    print("Buzz")
  else:
    print(i)
```

```python
units = int(input("Enter units consumed: "))

total_bill = 0

if units <= 100:
    total_bill = units * 5
elif units <= 200:
    total_bill = (100 * 5) + ((units - 100) * 7)
else:
    total_bill = (100 * 5) + (100 * 7) + ((units - 200) * 10)

print(f"Total Bill Amount: ₹{total_bill}")
```

```python
s = input("Enter num or string to be checked = ")

# Convert to string and remove spaces for accurate palindrome check
s_cleaned = str(s).replace(" ", "").lower()

if s_cleaned == s_cleaned[::-1]:
    print("Given num/string is Palindrome")
else:
    print("Given num/string is not Palindrome")
```
