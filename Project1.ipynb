Personal Expense Tracker

```python
# Personal Expense Tracker

expenses = []
monthly_budget = {
    "Food": 5000,
    "Transport": 2000,
    "Entertainment": 3000,
    "Utilities": 4000,
    "Others": 2000
}


def add_expense(date, category, amount):
    expense = {
        "date": date,
        "category": category,
        "amount": amount
    }
    expenses.append(expense)


def monthly_summary():
    total = 0
    for exp in expenses:
        total += exp["amount"]
    print("\nTotal Monthly Spending:", total)


def category_report():
    category_total = {}

    for exp in expenses:
        cat = exp["category"]
        amt = exp["amount"]

        if cat in category_total:
            category_total[cat] += amt
        else:
            category_total[cat] = amt

    print("\nCategory-wise Spending:")
    for cat, amt in category_total.items():
        print(cat, ":", amt)


        if cat in monthly_budget and amt > monthly_budget[cat]:
            print("âš  Budget exceeded for", cat)


def highest_spending_day():
    day_total = {}

    for exp in expenses:
        date = exp["date"]
        amt = exp["amount"]

        if date in day_total:
            day_total[date] += amt
        else:
            day_total[date] = amt

    highest_day = ""
    highest_amount = 0

    for day in day_total:
        if day_total[day] > highest_amount:
            highest_amount = day_total[day]
            highest_day = day

    print("\nHighest Spending Day:", highest_day, "Amount:", highest_amount)


def dominant_category():
    category_total = {}

    for exp in expenses:
        cat = exp["category"]
        amt = exp["amount"]

        if cat in category_total:
            category_total[cat] += amt
        else:
            category_total[cat] = amt

    dominant_cat = ""
    max_spend = 0

    for cat in category_total:
        if category_total[cat] > max_spend:
            max_spend = category_total[cat]
            dominant_cat = cat

    print("\nDominant Spending Category:", dominant_cat)


def high_spending_pattern(threshold=2000):
    print("\nHigh Spending Alerts:")
    for exp in expenses:
        if exp["amount"] > threshold:
            print("High expense of", exp["amount"], "on", exp["date"], "for", exp["category"])


add_expense("2026-01-01", "Food", 1200)
add_expense("2026-01-02", "Transport", 300)
add_expense("2026-01-02", "Entertainment", 2500)
add_expense("2026-01-03", "Food", 1800)
add_expense("2026-01-05", "Utilities", 4500)
add_expense("2026-01-05", "Others", 2200)


monthly_summary()
category_report()
highest_spending_day()
dominant_category()
high_spending_pattern()
```

Campus Store Inventory and Billing System

```python
# Inventory and Billing System for Campus Store

inventory = {
    "Pen": {"price": 10, "stock": 50},
    "Notebook": {"price": 50, "stock": 30},
    "Marker": {"price": 25, "stock": 20}
}

LOW_STOCK_LIMIT = 5
TAX_RATE = 0.05


def display_inventory():
    print("\nAvailable Products:")
    print("-" * 40)
    for item, details in inventory.items():
        print(f"{item} | Price: â‚¹{details['price']} | Stock: {details['stock']}")
        if details["stock"] <= LOW_STOCK_LIMIT:
            print("âš ï¸  Low Stock Alert!")
    print("-" * 40)


def admin_menu():
    while True:
        print("\n--- Admin Menu ---")
        print("1. Add Product")
        print("2. Update Stock")
        print("3. Update Price")
        print("4. View Inventory")
        print("5. Exit Admin Menu")

        choice = input("Enter choice: ")

        if choice == "1":
            name = input("Product name: ")
            if name in inventory:
                print("Product already exists.")
                continue
            price = float(input("Price: "))
            stock = int(input("Initial Stock: "))
            inventory[name] = {"price": price, "stock": stock}
            print("Product added successfully.")

        elif choice == "2":
            name = input("Product name: ")
            if name not in inventory:
                print("Product not found.")
                continue
            qty = int(input("Quantity to add: "))
            inventory[name]["stock"] += qty
            print("Stock updated.")

        elif choice == "3":
            name = input("Product name: ")
            if name not in inventory:
                print("Product not found.")
                continue
            price = float(input("New price: "))
            inventory[name]["price"] = price
            print("Price updated.")

        elif choice == "4":
            display_inventory()

        elif choice == "5":
            break

        else:
            print("Invalid choice.")


def calculate_discount(subtotal):
    if subtotal >= 2000:
        return 0.15
    elif subtotal >= 1000:
        return 0.10
    elif subtotal >= 500:
        return 0.05
    else:
        return 0.0


def customer_menu():
    cart = []

    while True:
        display_inventory()
        item = input("Enter product name to buy (or 'done' to checkout): ")

        if item.lower() == "done":
            break

        if item not in inventory:
            print("Product not available.")
            continue

        qty = int(input("Enter quantity: "))

        if qty <= 0:
            print("Invalid quantity.")
            continue

        if qty > inventory[item]["stock"]:
            print("Insufficient stock available.")
            continue

        cart.append({"item": item, "qty": qty})
        inventory[item]["stock"] -= qty
        print("Item added to cart.")

    generate_bill(cart)


def generate_bill(cart):
    if not cart:
        print("No items purchased.")
        return

    print("\nðŸ§¾ BILL")
    print("-" * 50)
    subtotal = 0

    for entry in cart:
        item = entry["item"]
        qty = entry["qty"]
        price = inventory[item]["price"]
        cost = qty * price
        subtotal += cost
        print(f"{item} | Qty: {qty} | Price: â‚¹{price} | Cost: â‚¹{cost}")

    discount_rate = calculate_discount(subtotal)
    discount = subtotal * discount_rate
    tax = (subtotal - discount) * TAX_RATE
    total = subtotal - discount + tax

    print("-" * 50)
    print(f"Subtotal: â‚¹{subtotal:.2f}")
    print(f"Discount ({int(discount_rate * 100)}%): -â‚¹{discount:.2f}")
    print(f"Tax (5%): â‚¹{tax:.2f}")
    print(f"Total Payable: â‚¹{total:.2f}")
    print("-" * 50)
    print("Thank you for shopping! ðŸ˜Š")


def main():
    while True:
        print("\n=== Campus Store System ===")
        print("1. Admin")
        print("2. Customer")
        print("3. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            admin_menu()
        elif choice == "2":
            customer_menu()
        elif choice == "3":
            print("Exiting system...")
            break
        else:
            print("Invalid choice.")


main()
```
